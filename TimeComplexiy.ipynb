```python
class Node:

    def __init__(self, v):
        self.val = v
        self.left = None
        self.right = None


def insert(root, node):
    if root is None:
        return node

    if root.val > node.val:
        root.left = insert(root.left, node)

    elif root.val < node.val:
        root.right = insert(root.right, node)

    return root


import random

randomlist1 = random.sample(range(3,7000),500 )
randomlist1 = list(set(randomlist1))
root = Node(0)
count=1

for i in randomlist1:
    count=count+1
    print(count)

    insert(root, Node(i))
print("data added successfully")

```

```python
import time
def BFS(goal):
    start_time = time.time()
    explored = []
    q = []
    q.append(root)


    while len(q)>0:

        #print(q)
        
        qq = [i.val for i in q]
        print(qq)
        
        node = q.pop(0)
        
        explored.append(node.val)
        
        print('POP: ',node.val)
        print()

        if node.val == goal:
            print('Result:  ',explored)
            break


        if node.left is not None:
            q.append(node.left)

        if node.right is not None:
            q.append(node.right)
    end_time = time.time()
    return end_time-start_time

```

```python

bfsTime=[]

res=BFS(randomlist1[100])
bfsTime.append(res)
res=BFS(randomlist1[300])
bfsTime.append(res)
res=BFS(randomlist1[400])
bfsTime.append(res)
res=BFS(randomlist1[100])
bfsTime.append(res)
res=BFS(randomlist1[200])
bfsTime.append(res)
```

```python


def DFS(goal):
    start_time = time.time()
    explored = []
    s = []
    s.append(root)

    while len(s) > 0:

        ss = [i.val for i in s]
        print(ss)

        node = s.pop()

        explored.append(node.val)



        if node.val == goal:
            print('Result:  ', explored)
            break

        if node.right is not None:
            s.append(node.right)

        if node.left is not None:
            s.append(node.left)
    end_time = time.time()
    return end_time - start_time
```
