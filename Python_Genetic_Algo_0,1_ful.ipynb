```python
import random 
import copy


POPULATION_SIZE = 8     
GENERATIONS = 10     
LEN_CHROMOSOME = 16 
split = 7
```

```python
def create_chrom():
    chrom = []
    for i in range(LEN_CHROMOSOME):
        if random.random()> 0.5:
            chrom.append(1)
        else:
            chrom.append(0)
    return chrom


def create_pop():
    pop = []
    for i in range(POPULATION_SIZE):
        a = create_chrom()
        pop.append(a)
    return pop

def my(x):
    return x[1]

def fitness(chrom):
    a=sum(chrom[ :3])
    b=sum(chrom[ :-4])
    add=a+b
    return(add)

def pop_with_fit(population):
    pop_fit = []
    for i in population:
        
        
        a = (i,fitness(i))
        
        pop_fit.append(a)
        
    return sorted(pop_fit, key=my, reverse=True)


def selection(pop):
    print(len(pop[:4]))
    return pop[:4]


def crosover(sel):
    cros_lis = []
    for i in range(0,len(sel),2):

        p1 = sel[i][0]
        p2 = sel[i+1][0]

        c1 = p1[:split]+p2[split:]
        c2 = p2[:split]+p1[split:]

        cros_lis.append(p1)
        cros_lis.append(p2)
        cros_lis.append(c1)
        cros_lis.append(c2)
        print(cros_lis)
    return cros_lis

def mutaion(crs):
    mut = []
    for a in crs:
        ind = random.randint(0,len(a)-1)
        if a[ind] == 1:
            a[ind] = 0
        else:
            a[ind] = 1
        mut.append(a)
    return mut

def show(pop,generation):
    print()
    print('Generation no [',generation,']', 'Best Chromosome: ',pop[0][0],'Fitness: ',pop[0][1])
    print(80*'-')
    for no,i in enumerate(pop):
        print('chromosome # ',no+1 ,'<<<',i[0],'>>>','Fitness:',i[1])
```

```python
Population = None

for i  in range(50):
    if Population is None :
        Population = create_pop()
        Popul = pop_with_fit(Population)
    else:
        
   
        show(Popul,i)
        Popul = selection(Popul)
        # print('selection',Popul)
        Popul = crosover(Popul)
        print('Crossover',Popul)
        Popul = mutaion(Popul)
        Popul = pop_with_fit(Popul)
    
        if (Popul[0][1] == 7):
            show(Popul,i)
            break
```
